### Ejercicio 14 Joyero 💍:
↪ Atributos: Materiales (oro/plata), técnica de engaste, valor de inventario, certificados de autenticidad, diseño propio y tienda.

```python
# --- PROGRAMA SIMULADOR DE ASISTENTE DE JOYERÍA (IA INTEGRADA) ---

# El diccionario/lista empieza vacío (se irán agregando las joyas)
joyero = []

# Función para mostrar de forma atractiva las joyas registradas en forma de tabla
def mostrar_joyas():
    if len(joyero) == 0:
        print("\n[Asistente] 📭 El joyero está vacío actualmente.")
        return

    print("\n" + "="*95)
    # Encabezados con formato de columnas alineadas
    print(f"{'Material':<12} | {'Técnica':<15} | {'Valor ($)':<12} | {'Certificado':<13} | {'Diseño Propio':<15} | {'Tienda Venta':<15}")
    print("="*95)

    # Ciclo for para recorrer la lista de diccionarios y mostrar la información bien distribuida
    for joya in joyero:
        print(f"{joya['Material']:<12} | {joya['Técnica']:<15} | {joya['Valor']:<12} | {joya['Certificado']:<13} | {joya['Diseño Propio']:<15} | {joya['Tienda']:<15}")
    print("="*95 + "\n")

# Función que procesa de manera "inteligente" una recomendación basada en el inventario
def asistente_ia():
    if len(joyero) == 0:
        print("\n🤖 [IA]: No tengo datos suficientes en el joyero para darte un reporte.")
        return

    print("\n🤖 [IA Analizando Inventario...]")
    total_joyas = len(joyero)
    oro_count = 0
    valor_total = 0

    for j in joyero:
        # Sumamos el valor del inventario (convertido a entero para operar)
        valor_total += int(j['Valor'])
        # Contamos cuántas joyas son de oro (ignorando mayúsculas/minúsculas)
        if j['Material'].lower() == "oro":
            oro_count += 1

    print(f"-> Actualmente gestionas {total_joyas} pieza(s) en total.")
    print(f"-> El valor acumulado en tu inventario es de: ${valor_total} USD.")

    # Elif e If para simular la toma de decisiones o diagnóstico de la IA
    if oro_count > (total_joyas / 2):
        print("💡 [Sugerencia IA]: Tu catálogo destaca en Alta Joyería (predomina el Oro). ¡Buen margen de ganancia!")
    elif oro_count == 0:
        print("💡 [Sugerencia IA]: Tu colección es puramente de Plata u otros materiales. Ideal para un público juvenil.")
    else:
        print("💡 [Sugerencia IA]: Tienes un inventario equilibrado entre Oro y Plata.")

# --- MENÚ INTERACTIVO PRINCIPAL ---
while True:
    print("✨ --- BIENVENIDO AL ASISTENTE VIRTUAL DE TU JOYERO --- ✨")
    print("1. Agregar una nueva joya")
    print("2. Ver inventario completo (Tabla)")
    print("3. Consultar al Asistente IA (Análisis de datos)")
    print("4. Salir del programa")

    opcion = input("Selecciona una opción (1-4): ")

    if opcion == "1":
        print("\n--- Registro de Nueva Joya ---")
        # Entrada de datos (key, values) provista por el usuario
        material = input("¿Material (Oro / Plata / Otro)?: ")
        tecnica = input("¿Técnica de engaste (Bisel / Pavé / Garras)?: ")
        valor = input("¿Valor de inventario (solo números enteros)?: ")
        certificado = input("¿Tiene certificado de autenticidad? (Sí / No): ")
        diseno = input("¿Es diseño propio? (Sí / No): ")
        tienda = input("¿En qué tienda se ubica asignada?: ")

        # Estructura del diccionario con los atributos solicitados
        nueva_joya = {
            "Material": material,
            "Técnica": tecnica,
            "Valor": valor,
            "Certificado": certificado,
            "Diseño Propio": diseno,
            "Tienda": tienda
        }

        # Agregamos el diccionario a nuestra lista 'joyero'
        joyero.append(nueva_joya)
        print("\n✅ ¡Joya guardada con éxito en el sistema!\n")

    elif opcion == "2":
        # Llamamos a la función encargada de dibujar la tabla
        mostrar_joyas()

    elif opcion == "3":
        # Llamamos a la función con lógica condicional que actúa como IA
        asistente_ia()

    elif opcion == "4":
        print("\n👋 ¡Gracias por usar el Asistente del Joyero Inteligente! Hasta pronto.")
        break  # Rompe el ciclo while True para finalizar

    else:
        print("\n❌ Opción no válida. Por favor, selecciona una opción del 1 al 4.\n")
print("Programa realizado por Angela Michelle Correa Lopez NL = 14")
```
